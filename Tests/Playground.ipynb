```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# limit the maximum derivative of the exponential function
# the exponential is linearized after a determined point
def exp_lim(x):
    return np.exp(x) if x < 70. else np.exp(70.) + np.exp(70.) * (x - 70.)

x = np.arange(65.9, 75.1, 0.001)
y = []
for i in x:
    y.append(exp_lim(i))

plt.plot(x, y)
plt.show()
```

```python
Is = 1e-15
N = 1.
Vt = 26e-3

Vdold = 0.
Vdn = np.arange(-2, 2, 0.01)
Vd = []
for _Vdn in Vdn:
    Vd.append(Vdold + 10. * N * Vt * np.tanh((_Vdn - Vdold) / (10. * N * Vt)))

plt.plot(Vdn, Vd)
plt.show()
```

```python
Is = 1e-15
N = 1.
Vt = 26e-3
Vcrit = N * Vt * np.log(N * Vt / (np.sqrt(2.) * Is))
Vdold = 0.
Vdn = np.arange(0, 5, 0.01)
Vd = []
for _Vdn in Vdn:
    if (_Vdn > 0. and _Vdn > Vcrit):
        Vd.append(Vdold +  N * Vt * np.log1p((_Vdn - Vdold) / (N * Vt)))
    else:
        Vd.append(_Vdn)

plt.plot(Vdn, Vd)
plt.show()
```
